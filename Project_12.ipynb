```python
# making a copy of dataset 
train_flight = train_flight_data.copy()
```

```python
# From date of Date_of_Journey column extracting day,month and weekdays respectively
train_flight['Journey_Day'] = pd.to_datetime(train_flight.Date_of_Journey, format='%d/%m/%Y').dt.day
train_flight['Journey_Month'] = pd.to_datetime(train_flight.Date_of_Journey, format='%d/%m/%Y').dt.month
train_flight['weekday']= pd.to_datetime(train_flight.Date_of_Journey, format='%d/%m/%Y').dt.weekday
```

```python
train_flight.columns
```

```python
train_flight['Total_Stops'].head()
```

```python
train_flight['Total_Stops'].unique()
```

```python
# converting 'Total_Stops' to 0, 1, 2, 3, 4 Stops respectively
train_flight['Total_Stops']=train_flight["Total_Stops"].replace({"non-stop":0, "1 stop":1, "2 stops":2, "3 stops":3, "4 stops":4})
```

```python
train_flight['Total_Stops'].unique()
```

```python
# Departue and arrival time  of flights, converting to morning, noon, evening, night and MidNight_EarlyMorning
def datime(f):
    f=f.strip()
    timetaken=(int)(f.split(':')[0])
    if(timetaken>=5 and timetaken<12):
        f='Morning'
    elif(timetaken>=12 and timetaken<16):
        f='Noon'
    elif(timetaken>=16 and timetaken<20):
        f='Evening'
    elif(timetaken>=20 and timetaken<24):
        f='Night'
    elif(timetaken>=0 and timetaken<5):
        f='MidNight_EarlyMorning'
    return f
train_flight['Dep_Time'] = train_flight['Dep_Time'].apply(datime)
train_flight['Arrival_Time'] = train_flight['Arrival_Time'].apply(datime)       
```

```python
# converting duration column into minutes
def duration(t):
    t = t.strip()
    tt = t.split(' ')
    td = tt[0]
    hrs = (int)(td[:-1])*60
    if((len(tt))==2):
        mint = (int)(tt[1][:-1])
        hrs = hrs + mint
    t = str(hrs)
    return t
train_flight['Duration'] = train_flight['Duration'].apply(duration)
```

```python
train_flight['Duration'].nunique()
```

```python
# renaming duration column to duration_minutes
train_flight.rename(columns={'Duration':'Duration_minutes'}, inplace=True)
```

```python
# Extracted data from column 'Date_of_Journey', hence dropping it.
# We have 'Total_Stops', hence dropping column 'Route'
train_flight.drop(['Date_of_Journey','Route'], inplace=True, axis=1)
```

```python
# importing important libraries

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# finding best random state between 42 to 100 with test_size=0.22
max_r_score = 0
for r_state in range(42,101):
    x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = r_state, test_size=0.22)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    y_predict = lr.predict(x_test)
    r2_scr = r2_score(y_test, y_predict)
    print("r2_score corresponding to random state: ",r_state," is: ", r2_scr)
    if r2_scr > max_r_score:
        max_r_score = r2_scr
        final_r_state = r_state
    print()
    print()
    print("Maximum r2 score corresponding to",final_r_state, "is ",max_r_score)
```

```python
# since the maximun r2 score is comming around 0.547 with random state 56. As the score is very low 
# hence checking with other models
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 54, test_size = 0.22)
```

```python
y_train.shape
```

```python
# finding the best parameters for each model using gridsearchCV
```

```python
svr=SVR()
parameters = {'kernel':('linear','poly','rbf'), 'C':[1,10]}
svgs = GridSearchCV(svr, parameters, cv = 5)
svgs.fit(x_train,y_train)
svgs.best_params_
```

```python
knn = KNeighborsRegressor()
parameters = {'n_neighbors':range(1,30)}
knngs = GridSearchCV(knn, parameters, cv = 5)
knngs.fit(x_train, y_train)
knngs.best_params_
```

```python
rfr = RandomForestRegressor()
parameters = {'n_estimators':[50,100,150,200,250,300]}
rfgs = GridSearchCV(rfr, parameters, cv = 5)
rfgs.fit(x_train, y_train)
rfgs.best_params_
```

```python
dtr = DecisionTreeRegressor()
parameters = {'criterion':('mse','friendman_mse', 'mse')}
dtgs = GridSearchCV(dtr, parameters, cv = 5)
dtgs.fit(x_train, y_train)
dtgs.best_params_
```

```python
lso = Lasso()
parameters = {'alpha':[0.001,0.01,0.1,1]}
lsgs = GridSearchCV(lso, parameters, cv = 5)
lsgs.fit(x_train, y_train)
lsgs.best_params_
```

```python
gbr = GradientBoostingRegressor()
parameters = {'learning_rate':[0.001,0.01,0.1,1], 'n_estimators':[10,100,500,1000]}
gbgs = GridSearchCV(gbr, parameters, cv = 5)
gbgs.fit(x_train, y_train)
gbgs.best_params_
```

```python
lr = LinearRegression()
svr = SVR(C=10, kernel = 'linear')
knn = KNeighborsRegressor(n_neighbors = 5)
rfr = RandomForestRegressor(n_estimators = 200)
dtr = DecisionTreeRegressor(criterion ='mse')
lso = Lasso(alpha =1 )
gbr = GradientBoostingRegressor(learning_rate = 1, n_estimators = 1000)
```

```python
models = []
models.append(('LinearRegression', lr))
models.append(('SVR', svr))
models.append(('KNeighborsRegressor', knn))
models.append(('RandomForestRegressor', rfr))
models.append(('DecisionTreeRegressor', dtr))
models.append(('Lasso', lso))
models.append(('GradientBoostingRegressor', gbr))
```

```python
# Testing all required models with best parameters calculated

Model = []
score = []
rmse = []
mae = []
mse = []
cvs = []
std = []

for name,model in models:
    print ('******************', name ,'********************')
    print("\n")
    Model.append(name)
    model.fit(x_train, y_train)
    print(model)
    pred = model.predict(x_test)
    print("\n")
    rscore = r2_score(y_test, pred)
    print("r2 Score =", rscore)
    score.append(rscore*100)
    print("\n")
    RMS = np.sqrt(mean_squared_error(y_test, pred))
    print("RMSE =", RMS)
    rmse.append(RMS*100)
    print("\n")
    MA = mean_absolute_error(y_test, pred)
    print("MAE =", MA)
    mae.append(MA*100)
    print("\n")
    MS = mean_squared_error(y_test, pred)
    print("MSE =", MS)
    mse.append(MS*100)
    print("\n")
    cvscore = cross_val_score(model, x, y, cv = 10).mean()
    print("Cross_Val_Score =", cvscore)
    cvs.append(cvscore*100)
    print("\n")
    stdscore = cross_val_score(model, x, y, cv = 10).std()
    print("Standard Deviation =", stdscore)
    std.append(stdscore*100)
    print("\n\n")
```

```python
import pandas as pd
result = pd.DataFrame({'Model': Model, 'r2_score': score, 'Root_mean_square': rmse, 'Mean_absolute_error': mse, 'Mean_square_error': mse, 'Cross_val_score_mean': cvs, 'Cross_val_score_Std': std})
result
```
